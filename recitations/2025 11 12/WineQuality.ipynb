```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('WineQT.csv')
```

```python
df.head()
```

```python
df["citric acid"].replace(0.00,np.nan, inplace=True)
```

```python
df.head()
```

```python
# TODO 1.	Fill all null values in citric acid column with mean value of the column
df['citric acid'] = df['citric acid'].fillna(df['citric acid'].mean())
# df.loc[df['citric acid'].isna(), 'citric acid'] = df['citric acid'].mean()
```

```python
sum(df['citric acid'].isna())
```

```python
# TODO 2.	Using iloc - Print “alcohol” values for the first 10 rows
df.iloc[:10, list(df.columns).index('alcohol')]
```

```python
# TODO 3.	Using loc - Select rows where alcohol > 10 AND show only alcohol and quality columns
df.loc[df.alcohol > 10, ['alcohol', 'quality']]
```

```python
# TODO 4.	Create a new dataframe such that you want to know the average alcohol content , minimum fixed acidity and maximum fixed acidity by the quality rating of wine (use group by)
agg = df.groupby('quality').aggregate({'alcohol': 'mean', 'fixed acidity': ['min', 'max']})
agg
```

```python
agg.columns
```

```python
# TODO 5. Sort the values by alcohol mean
agg.sort_values(('alcohol', 'mean'))
```

```python
#TODO 6. Do the same using pivot table
df.pivot_table(index='quality', aggfunc={'alcohol': 'mean', 'fixed acidity': ['min', 'max']})
```
